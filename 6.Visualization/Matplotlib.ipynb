# Basic Plot

```python
x=np.arange(1,10)
y=x**2
print(x)
print(y)
```

```python
plt.plot(x,y,"r",marker="o",linestyle=":",linewidth=2,markersize=10)
plt.title("Line Plot")
plt.xlabel("Number")
plt.ylabel("Square")
```

## Custmizing the Height and Width

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,"r",marker="o",linestyle=":",linewidth=2,markersize=10)
plt.title("Line Plot")
plt.xlabel("Number")
plt.ylabel("Square")
```

# Box Plot
##### 1.It is standardized way to display the distribution of quantative data (min,Q1,Q2,Q3,max)
##### 2.Edges of box is IQR=Q3-Q1
##### 3.It needs one numberical column to create the box plot

```python
df=pd.read_csv("cars93.csv")
df.head()
```

```python
fig=plt.figure(figsize=(6,4))
plt.boxplot(df["Price"])
plt.title("Box Plot")
plt.show()
```

```python
fig=plt.figure(figsize=(4,2))
plt.boxplot(df["Passengers"])
plt.title("Box Plot")
plt.show()
```

```python
fig=plt.figure(figsize=(4,2))
plt.boxplot(df["MPG.city"])
plt.title("Box Plot")
plt.show()
```

### To get the data of box plot 

```python
B=plt.boxplot(df["Price"])
plt.title("Box Plot")
plt.xlabel("Values")
plt.ylabel("Price")
[item.get_ydata()[0] for item in B["whiskers"]]#min and max value
```

### Conclusion:
 It has 3 outliers.It's min value is 2501000 and max value is 4776500.

# Scatter Plot
#### 1.It represent values for two different numerical columns
#### 2.It is used to observing the relationship between two numerical columns

```python
plt.scatter(df["Price"],df["Horsepower"],color="r")
plt.title("Scatter Plot")
plt.xlabel("Price")
plt.ylabel("Horsepower")
plt.show()
```

### Conclusion:
  As Price increases Housepower is also increases .It is directly proportional to each other.

# Histogram
#### 1.Representation of the distribution of numerical data.
#### 2. It give as is column follows normal distribution or not.

```python
plt.hist(df["Horsepower"])
plt.title("Histogram of Horsepower")
plt.ylabel("Horsepower")
plt.show()
```

## Conclusion:
Horsepower does not follow normal distribution bcoz the shape of the histogram does not bell curve.

```python
#Using bin size plot histogram
plt.hist(df["Horsepower"], bins = 20)
plt.title("Histogram of Horsepower")
plt.ylabel("Horsepower")
plt.show()
```

# Bar Chart
#### 1. It contain the caterical data.
#### 2.It used to comparisons of discrete categories.

```python
df["Type"].value_counts()
```

```python
df["Type"].value_counts().index
```

```python
df["Type"].value_counts().values
```

```python
plt.bar(df["Type"].value_counts().index,df["Type"].value_counts().values,color="r",linestyle=":",width=0.8,align="center",edgecolor="Blue",linewidth=2,hatch="/")
plt.xlabel("Type")
plt.ylabel("Counts")
plt.title("Distribution of Types across various categorices")
plt.show()
```

## Conclusion:
Midsize has max count and Van is min count

# Line Chart
#### 1.To find relationship between two numerical column or to visualize a trend in time series data
#### 2. Points are unordered 

```python
plt.plot(df["Horsepower"],df["MPG.city"])
```

```python
ordered=df.sort_values(by="Horsepower",ascending=False)
ordered
ordered.head()
```

```python
plt.plot(ordered["Horsepower"],ordered["MPG.city"])
plt.title("Line Chart")
plt.xlabel("Horsepower")
plt.ylabel("MPG.city")
plt.show()
```
